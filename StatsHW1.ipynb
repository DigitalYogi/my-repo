```python
# What is the average for points scored in wins and losses?

# Wins Average
wins = nba[(nba['game_result'] == "W")]
print("Average points for Wins:", int(wins.pts.mean()))

# Loses Average
loses = nba[(nba['game_result'] == "L")]
print("Average points for Loses:", int(loses.pts.mean()))
```

```python
# What is the average points scored for the Spurs? (Hint SAS is the team id)

spurs = nba[(nba['team_id'] == "SAS")]
print("Average point scored by the spurs?:", round(spurs.pts.mean()))
```

```python
# What is the median difference between Team Score and Opponent Score?

# Team 1
team_med = nba['pts'].median()
print("Team median: ", team_med)

# Opposing team
opp_med = nba['pts'].median()
print("Opponent median: ", opp_med)

# Median difference
print("Median difference: ", abs(team_med - opp_med))
```

```python
# How has the average and median points scored for the winning and losing teams per both teams changed by decade?
#40,s, 50's, 60's, 70's, 80's, 90's, 2000's, 2010's

years = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
wins = nba[nba["game_result"] == 'W']
ls = nba[nba["game_result"] == 'L']

for i in years:
  decade_w = wins[(wins["year_id"]>=i) & (wins["year_id"]<i+10)]
  decade_l = ls[(ls["year_id"]>i) & (ls['year_id']<i+10)]

  print("Mean points scored by winning team in", i, "was", decade_w.pts.mean())
  print("Mean points scored by lossing team in", i, "was", decade_l.pts.mean())
  print("")
  print("Median points scored by winning team in", i, "was", decade_w.pts.median())
  print("Median points scored by lossing team in", i, "was", decade_l.pts.median())
```
